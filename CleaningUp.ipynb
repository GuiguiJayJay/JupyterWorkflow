# Unsupervised analysis of days of the week

Treating crossings each day as features to learn about the relationship between various days

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')

import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
```

## Get the data

```python
from jupyterworkflow.data import get_freemont_data
data = get_freemont_data()
```

```python
pivoted = data.pivot_table('Total', index=data.index.time, columns=data.index.date)
pivoted.plot(legend=False, alpha=0.01)
```

## PCA

```python
X = pivoted.fillna(0).T.values
X.shape
```

```python
X2 = PCA(2, svd_solver='full').fit_transform(X)
X2.shape
```

```python
plt.scatter(X2[:,0], X2[:,1])
```

## Unsupervised Clustering

```python
gmm = GaussianMixture(2).fit(X)
labels = gmm.predict(X)
```

```python
plt.scatter(X2[:,0], X2[:,1], c=labels, cmap='rainbow')
plt.colorbar()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

pivoted.T[labels == 0].T.plot(legend=False, alpha=0.1, ax=ax[0])
pivoted.T[labels == 1].T.plot(legend=False, alpha=0.1, ax=ax[1])

ax[0].set_title('Red cluster')
ax[1].set_title('Purple cluster')
```

## Comparing with day of week

```python
daysofweek = pd.DatetimeIndex(pivoted.columns).dayofweek
```

```python
plt.scatter(X2[:,0], X2[:,1], c=daysofweek, cmap='rainbow')
plt.colorbar()
```
